```python
from irt import irt

from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
from scipy import optimize
%matplotlib inline
```

```python
thetas, betas, answered, correct = irt.simulate(100, 30, 15)
```

```python
plt.hist(thetas)
```

```python
plt.hist(betas)
```

```python
f = irt.make_irt_neg_objective(answered, correct)
output = optimize.minimize(f, 1.0 - 2.0 * np.random.random(thetas.size + betas.size), method='L-BFGS-B')
```

```python
values = output['x']
rec_thetas = values[:thetas.size]
rec_betas = values[thetas.size:]
```

```python
sns.jointplot(thetas, rec_thetas)
```

```python
sns.jointplot(betas, rec_betas)
```
