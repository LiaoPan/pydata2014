```python
import numpy as np
from scipy import optimize

from football import base, elo
```

```python
# Setup data
games = base.get_games(open('cfb2013lines.csv', 'r'))
conferences = base.get_teams_to_conference(open('conferences.csv', 'r'))
edges = map(base.win_or_lose, games)
team_to_idx, int_winners, int_losers = elo.convert_edges(edges)
```

```python
# Make likelihood
likelihood = elo.make_elo_neg_likelihood(int_winners, int_losers)
```

```python
# Let scipy do the hard work
output = optimize.minimize(likelihood, np.zeros(max(np.max(int_winners), np.max(int_losers)) + 1), method='L-BFGS-B')
```

```python
# Match indices to teams
values = output['x']
idx_to_team = sorted((v, k) for k, v in team_to_idx.viewitems())
idx_to_team = [v for _, v in idx_to_team]
```

```python
# Get sorted order
value_and_team = [(values[i], idx_to_team[i]) for i in range(len(idx_to_team))]
value_and_team.sort()
value_and_team.reverse()
```
